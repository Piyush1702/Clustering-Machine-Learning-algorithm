```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.rcParams['figure.figsize']=(12,7)
plt.style.use('ggplot')
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset.head(10)
```

```python
dataset['Genre'].value_counts()
```

```python
f1=dataset['Annual Income (k$)'].values
f2=dataset['Spending Score (1-100)'].values

for keys in dataset.keys():
    print(keys)
    
X=np.array(list(zip(f1, f2)))

plt.scatter(f1, f2, c='black', s=20)
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
labels=kmeans.predict(X)
```

```python
c= kmeans.cluster_centers_
c
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=Axes3D(fig)

ax.scatter(X[:,0],X[:,1], X[:,1])
```

```python
fig=plt.figure()
ax= Axes3D(fig)
ax.scatter(X[:,0],X[:,1], X[:,1], c='b')
ax.scatter(c[:,0],c[:,1], c[:,1], marker='*',c='#050505',s=1000)
```
